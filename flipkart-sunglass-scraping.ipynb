```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.flipkart.com/search?q=sunglass&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
```

```python
page = requests.get(url)
page
```

```python
soup=BeautifulSoup(page.content, 'html.parser')
soup
```

```python
brand=[]
description=[]
price=[]
```

```python
names=soup.find_all('div',class_ ="syl9yP")
for i in names:
    name=i.text
    brand.append(name)
brand=brand[0:20]
print(brand)
len(brand)
```

```python
desc = soup.find_all('a',class_='WKTcLC BwBZTg')
for i in desc:
    name=i.text
    description.append(name)
description=description[0:20]
print(description)
len(description)
```

```python
prices = soup.find_all('div',class_='Nx9bqj')
for i in prices:
    name=i.text
    price.append(name)
price=price[0:20]
print(price)
len(price)
```

```python
# Creating Dataframe
df=pd.DataFrame({"Brand Name" : brand,"Description" : description,"Prices" : price})

print(df)
```

```python
df["Prices"] = df["Prices"].str.replace("₹", "", regex=False)
```

```python
print(df)
```

```python
df.to_csv("flipkart_sunglasses_dataset.csv",index=False)
```
